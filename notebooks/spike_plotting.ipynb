```python
from norse.torch.module.lif import LIFFeedForwardCell
import torch
import numpy as np
```

```python
lif = LIFFeedForwardCell([10])
data_in = np.array([np.zeros(100) + i for i in np.arange(0, 2, 0.2)]).T
s0 = lif.initial_state(1, "cpu")
```

```python
spikes = []
state = s0
for data in data_in:
    out, state = lif(data, state)
    spikes.append(np.array(out[0]))
spikes = np.array(spikes)
```

```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
spike_data = [[x for x, spike in enumerate(inner_spikes) if spike > 0] for inner_spikes in spikes.T]
plt.figure(figsize=(10, 10))
plt.xlabel("Timestep")
plt.ylabel("Neuron index")
for i in range(10):
    plt.eventplot(spike_data)
```
